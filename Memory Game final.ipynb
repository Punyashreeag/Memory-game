```python
#final
import cv2
import numpy as np

#drawing rectangles
box=np.zeros((430,640,3),np.uint8)
cv2.rectangle(box,(10,10),(210,210),(150,255,200),-1)
cv2.rectangle(box,(220,10),(420,210),(150,255,200),-1)
cv2.rectangle(box,(430,10),(630,210),(150,255,200),-1)
cv2.rectangle(box,(10,220),(210,420),(150,255,200),-1)
cv2.rectangle(box,(220,220),(420,420),(150,255,200),-1)
cv2.rectangle(box,(430,220),(630,420),(150,255,200),-1)


#finding contours



gray=cv2.cvtColor(box,cv2.COLOR_BGR2GRAY)
cnts,hierarchy = cv2.findContours(gray, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)

#on mouse click
global moves
moves=0
global prev
prev=10
global posprev
posprev=10
global right
right=0
global current
current=20
global pos
pos=10

def play(event,x,y,flags,param):
    if event==cv2.EVENT_LBUTTONDBLCLK:
        count=0
        global moves,prev,posprev,right,current,pos
        if(moves%2==0 and moves!=0):
            if(prev!=current and posprev!=prev):
                if(pos==0 or posprev==0):
                    box[220:420,430:630]=[150,255,200]
                if(pos==1 or posprev==1):
                    box[220:420,220:420]=[150,255,200]
                if(pos==2 or posprev==2):
                    box[220:420,10:210]=[150,255,200]
                if(pos==3 or posprev==3):
                    box[10:210,430:630]=[150,255,200]
                if(pos==4 or posprev==4):
                    box[10:210,220:420]=[150,255,200]
                if(pos==5 or posprev==5):
                    box[10:210,10:210]=[150,255,200]  
                
        prev=current
        posprev=pos
                    
        for c in cnts:
            dist = cv2.pointPolygonTest(cnts[count],(x,y),True)
            if(dist>=0):
                if (count==0):
                    box[220:420,430:630]=panda
                    current=3
                    pos=0
                    if(pos!=posprev):
                        moves=moves+1
                   
                elif(count==1):
                    box[220:420,220:420]=lion
                    current=1
                    pos=1
                    if(pos!=posprev):
                        moves=moves+1
                  
                  
                    
                elif(count==2):
                    box[220:420,10:210]=tiger
                    current=2
                    pos=2
                    if(pos!=posprev):
                        moves=moves+1
                    
                    
                elif(count==3):
                    box[10:210,430:630]=tiger
                    current=2
                    pos=3
                    if(pos!=posprev):
                        moves=moves+1
                   
                      
                elif(count==4):
                    box[10:210,220:420]=panda
                    current=3
                    pos=4
                    if(pos!=posprev):
                        moves=moves+1
                   
                    
                    
                elif(count==5):
                    box[10:210,10:210]=lion
                    current=1
                    pos=5
                    if(pos!=posprev):
                        moves=moves+1
                  
                 
                else:
                    continue
                   
                    
                    
            count=count+1

        if(moves%2==0):
            if(prev==current):
                right=right+1
                if(right==3):
                    cv2.putText(box, "GAME COMPLETE", (205, 205),  cv2.FONT_HERSHEY_SIMPLEX, 0.5,(200, 200, 0), 2)
            
                    
        
                        
                    
            
        
            
            

    
 #images   

lion=cv2.imread('lion.jpg')
lion=cv2.resize(lion,(200,200))


tiger=cv2.imread('tiger.jpg')
tiger=cv2.resize(tiger,(200,200))

panda=cv2.imread('panda2.jpg')
panda=cv2.resize(panda,(200,200))



cv2.namedWindow('image')
       
while(1):
   
    cv2.setMouseCallback('image',play)
    cv2.imshow('image',box)
   
    if cv2.waitKey(20) & 0xFF==27:
        break
cv2.destroyAllWindows()   
```
